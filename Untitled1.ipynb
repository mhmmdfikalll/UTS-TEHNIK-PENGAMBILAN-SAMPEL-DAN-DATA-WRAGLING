```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
base_url = "https://myhartono.com/page/"
data = []

for page in range(1, 4):
    url = base_url + str(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # ambil semua produk (ubah sesuai struktur web)
    products = soup.find_all("div", class_="product-item")

    for p in products:
        title = p.find("a", class_="product-title")
        price = p.find("span", class_="price")

        if title and price:
            data.append({
                "nama": title.text.strip(),
                "harga": price.text.strip()
            })
```

```python
with open("data_lis.json", "w") as f:
    json.dump(data, f, indent=4)
```

```python
df = pd.DataFrame(data)
df.to_csv("data_lis.csv", index=False)
```
